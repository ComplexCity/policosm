```python
wiki = """{{Languages|OSM tags for routing/Access Restrictions}}

[[Relations/Proposed/Defaults|Default]] is a '''proposal''' (not yet fully drafted, no RFC, no voting yet) for a possible, upcoming default value system. This proposal may set default access restrictions by area (country, state...) if it becomes a generally accepted practice. 

The access restrictions, or allowance if no restriction is implied, concern the '''legal''' access and not the practical access in terms of convenience. When legal access exists, but practical access is not, this can either be solved by adding surface, gradient and smoothness tags. This to advise e.g. wheelchairs, racing bikes etc. and routing algorithms not to use bad, unpaved, soft and steep paths.

Note that explicit tagging of access repeating what following tables imply may be still useful. For example bare {{tag|highway|path}} may mean either "surveyed and all defaults apply" or "mapped from LIDAR, not surveyed, may be private". In contrast {{tag|highway|path}} + {{tag|foot|yes}} + {{tag|bicycle|yes}} is more clear what is its meaning.

==== Interpretation ====

* '''yes''' - this kind of vehicle is allowed to drive on this kind of road unless tagged otherwise on the individual way (e.g. "motorcar=no").
* '''designated''' - same as '''yes''' but this road can optionally be preferred by some of your metrics.
* '''no''' - this kind of vehicle is not allowed to drive on this kind of road unless tagged otherwise on the individual way (e.g. "motorcar=yes").

See [[Key:access]] for the way such limits are usually tagged on the individual way.

==== Default ====

Default access restrictions for all countries.

Note: ''italic'' values are inherited from the generic access value (for instance, a specific cycleway with access=destination implies motorcar=destination, not motorcar=no).

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| access
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

 |-
 | motorway
motorway_link
 | {{yes}}
 | {{yes|Designated}}
 | {{yes|Designated}}
 | {{yes|Designated}}
 | {{yes|Designated}}
 | {{yes|Designated}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | trunk
trunk_link
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | primary

primary_link

secondary

secondary_link

tertiary

tertiary_link

unclassified

residential

living_street

road

 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 |-
 | pedestrian
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}

 |-
 | path 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|Designated}}
 | {{no}}
 | {{no}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|Designated}}
 | {{no}}

 |-
 | footway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|Designated}}
 |}

==== Australia ====

Proposed access restrictions. Not endorsed by the Australian OSM community (yet).

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| access
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

 |-
 | motorway
 | {{yes}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | trunk
 | {{yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}

 |-
 | primary

secondary

tertiary

unclassified

residential

living_street

road

 | {{yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 |-
 | pedestrian
 | {{no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{yes|yes}}
 | {{yes|designated}}

 |-
 | path 
 | {{no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway 
 | {{no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{yes|designated}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}

 |-
 | cycleway 
 | {{no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{yes|designated}}
 | {{yes|designated}}

 |-
 | footway 
 | {{no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{yes|designated}}
 |}

==== Austria ====

Default access restrictions if not tagged on the individual street.

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot
!| inline_skates

 |-
 | motorway
trunk
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{partial|yes}} (1)
 | {{yes}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | primary

secondary

tertiary

unclassified

residential

road
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{partial|yes}} (1)
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}} (2)
 | {{yes}} (3)
 | {{yes}} (3)

 |-
 | living_street (9)
 | {{partial|destination}}
 | {{partial|destination}}
 | {{partial|destination}}
 | {{partial|destination}} (1)
 | {{partial|destination}}
 | {{partial|destination}}
 | {{yes}}
 | {{yes}} (4)
 | {{yes}}
 | {{yes}}

 |- 
 | service
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}} (1?)
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | track (5)
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}} (1?)
 | {{yes}}
 | {{yes}}
 | {{yes}} (6)
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | path 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no|dismount}} (7)
 | {{yes}}
 | {{yes}}

 |-
 | bridleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{no}}
 | {{yes}}

 |-
 | footway 
pedestrian
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no|dismount}} (8)
 | {{yes|designated}}
 | {{yes}}

 |}
'''Notes:'''
# Forbidden on Sat 15:00-24:00 and Sun+holidays 0:00-22:00, with exceptions like food or cattle transports. Forbidden at 22:00-05:00 when maximum permissive gross weight exceeds 7.5 t, except for certified low-noise vehicles. See [http://www.ris.bka.gv.at/Dokument.wxe?Abfrage=Bundesnormen&Dokumentnummer=NOR40123906 StVO §42] for details.
# Should use nearby cycleway (see {{tag|bicycle|use_sidepath}}) if present, but road racing bikes are excepted from explicit or implicit bicycle=use_sidepath.
# Should use nearby sidewalk ({{tag|foot|use_sidepath}}) if present.
# {{tag|highway|living_street}} also implies {{tag|oneway:bicycle|no}}.
# Access on tracks is not restricted unless prohibited by some sign or barrier.
# On designated forest roads ({{tag|highway|track}} + {{tag|vehicle|forestry}}), {{key|horse||no}} is also implied due to [http://www.ris.bka.gv.at/Dokument.wxe?Abfrage=Bundesnormen&Dokumentnummer=NOR12132163 Forstgesetz §33] Abs. 3.
# StVO does not apply to paths. Therefore, other laws such as [http://www.ris.bka.gv.at/Dokument.wxe?Abfrage=Bundesnormen&Dokumentnummer=NOR12018065 ABGB §339] or [http://www.ris.bka.gv.at/Dokument.wxe?Abfrage=Bundesnormen&Dokumentnummer=NOR12132163 Forstgesetz §33] Abs. 3 are in force which generally do not differentiate between motorized and non-motorized vehicles.
# Dismounted cyclists count as pedestrians due to [https://www.ris.bka.gv.at/Dokument.wxe?Abfrage=Bundesnormen&Dokumentnummer=NOR40065117 StVO §65] Abs. 1.
#Both ''Wohnstraße'' and ''Begegnungszone'', two legally different road types, are typically mapped as {{Tag|highway|living_street}}. The described access defaults are actually only correct for a ''Wohnstraße''. For a ''Begegnungszone'' the differences should be mapped explicitly as described in [[DE:Tag:highway=living street]].

==== Belarus ====

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

 |-
 | motorway
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | trunk
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{partial|yes(1)}}
 | {{yes}}

 |-
 | primary

secondary

tertiary

unclassified

residential

 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{partial|yes}}
 | {{yes}}

 |-
 |living_street
 | {{yes|destination}}
 | {{yes|destination}}
 | {{yes|destination}}
 | {{yes|destination}}
 | {{yes|destination}}
 | {{yes|destination}}
 | {{yes}}
 | {{partial|yes}}
 | {{yes}}

 |-
 | path 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{partial|yes (2)}}
 |-
 | footway 

 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes|designated}}

 |-
 | pedestrian

 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes|designated}}

 |}

# Should use nearby cycleway, footway or sidewalk if present.
# Should use nearby footway if present. Most often is specifically designated for both, though, and tagged as such. If only a cycleway is present next to a road without sidewalks (never seen such a case), pedestrians "should" use the cycleway.

==== Belgium ====

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| motorcar
!| motorcycle
!| goods=hgv
!| bus
!| taxi
!| agricultural
!| moped
!mofa
!| horse
!| bicycle
!| foot

 |-
 | motorway
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | trunk
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | primary

secondary

tertiary

unclassified

residential

living_street

 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | track (proposal) (5)
 | {{partial}}
 | {{partial}}
 | {{partial}}
 | {{partial}}
 | {{partial}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 |-
 | path 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{no}}
 | {{yes|yes (1)}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|yes (4)}}
 | {{yes|yes (4)}}
 | {{no}}
 | {{yes}}
 | {{yes|yes (2)}}

 |-
 | footway 

 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}

 |-
 | pedestrian
 | {{no}} (6)
 | {{no}} (6)
 | {{no}} (6)
 | {{no}} (6)
 | {{no}} (6)
 | {{no}} (6)
 | {{no}} (6)
 | {{no}} (6)
 | {{no}} (6)
 | {{partial}} (3)
 | {{yes}}

 |}

'''Notes:'''
* designated = yes: There's no reason for a "designated" access tag in Belgium as there is no reason why one has more rights over the other on any of these highway types when different vehicle types have access to a road. "designated" is therefore synonym with "yes".  Footways could both be signed with a sign that doesn't show a pedestrian at all, and one that does, so basing a designated tag on traffic signs is also flawed.
* There's no concept of "hgv", so it's a synonym with "goods"
* psv is meaningless in Belgium and replaced by bus and taxi

# this assumes a bridleway that doesn't belong to a road. When it does belong to a road it depends on whether there's another place pedestrians can walk.
# cycleways allow pedestrians if there is no parallel path where pedestrians can go, it's safe to assume pedestrians are allowed
# Bicycles are often allowed on pedestrian roads, but this depends on the sign used. Sometimes it's only allowed during certain times of the day. However, all vehicles (including bikes) are required to drive at walking speed, and bikers must dismount when the density of walkers requires it. So having access or not doesn't matter a lot for routers.
# Moped/mofa in Belgium actually comprises three vehicle classes: Moped Class A (called {{Tag|mofa}}), Moped Class B (called {{Tag|moped}}) and Moped Class P (called {{Tag|speed_pedelec}}). Cycleways may allow any combination of them (including none at all), depending on the signs used, and the road the cycleway belongs to.
# on tracks, there are no national restrictions. Although the surface may limit certain vehicle types. Some cities also forbid the usage of tracks for motorcycles and/or quadbikes.
# About all vehicles are allowed to use pedestrian areas if it's absolutely necessary. F.e. to reach the buildings in the pedestrian area if it's the only possible way. However, a permit from the mayor is needed before usage is legal. As such, it's best for routers to consider pedestrian areas as non-accessible for all motorised vehicles. Many of the pedestrian areas also have sub-signs that allow for delivery on certain times of the day, even without permit.

==== Brazil ====

Default access restrictions according to the (pt) [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm Brazilian Transit Code] if not tagged on the individual street.

{| class="wikitable"
!| {{TagKey|highway}}
!| access
!| emergency<sup>1</sup>
!| motorcar<br/>motorcycle
!| goods<br/>hgv<br/>psv
!| agricultural
!| moped<sup>2</sup>
!| horse<sup>2</sup>
!| bicycle<br/>mofa<sup>3</sup>
!| foot<br/>inline_skates<sup>4</sup>
 |-
 | motorway<sup>2,5</sup>
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}<sup>6</sup>
 | {{no}}<sup>7,8</sup>
 | {{no}}<sup>5,6</sup>
 | {{partial|yes}}<sup>5, 8, 20</sup>
 | {{partial|yes}}<sup>5, 9, 20</sup>

 |-
 | trunk<sup>10</sup><br/>primary<sup>10</sup><br/>secondary<sup>10</sup><br/>tertiary<sup>10</sup><br/>unclassified<sup>10</sup><br/>road
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}<sup>6</sup>
 | {{yes}}<sup>7</sup>
 | {{yes}}<sup>11</sup>
 | {{yes}}<sup>8,12</sup>
 | {{yes}}<sup>9</sup>
 |-
 | residential
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}<sup>7</sup>
 | {{yes}}<sup>11</sup>
 | {{yes}}<sup>12</sup>
 | {{yes}}
 |-
 | living_street<sup>13</sup>
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}<sup>13</sup>
 | {{yes}}<sup>13</sup>
 | {{yes}}<sup>7</sup>
 | {{yes}}<sup>11</sup>
 | {{yes}}<sup>12</sup>
 | {{yes}}
 |-
 | service<sup>14</sup>
 | {{partial|destination}}
 | {{partial|destination}}<sup>15</sup>
 | {{partial|destination}}
 | {{partial|destination}}
 | {{partial|destination}}
 | {{partial|destination}}
 | {{partial|destination}}
 | {{partial|destination}}
 | {{partial|destination}}
 |-
 | track<sup>16</sup>
 | {{partial|destination}}
 | {{partial|discouraged}}<sup>17</sup>
 | {{partial|destination}}
 | {{partial|destination}}
 | {{partial|destination}}
 | {{partial|destination}}
 | {{partial|destination}}
 | {{partial|destination}}
 | {{partial|destination}}
 |-
 | pedestrian
 | {{yes}}
 | {{partial|discouraged}}<sup>18</sup>
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}<sup>7</sup>
 | {{no}}<sup>11</sup>
 | {{partial|dismount}}<sup>19</sup>
 | {{yes}}<sup>20</sup>
 |-
 | footway
 | {{yes}}
 | {{no}}<sup>21</sup>
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}<sup>7</sup>
 | {{no}}<sup>11</sup>
 | {{partial|dismount}}<sup>19</sup>
 | {{yes}}<sup>20</sup>
 |-
 | cycleway<sup>22</sup>
 | {{yes}}
 | {{no}}<sup>21</sup>
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}<sup>7</sup>
 | {{no}}<sup>11</sup>
 | {{yes|designated}}<sup>22</sup>
 | {{no}}<sup>23</sup>
 |-
 | bridleway<sup>24</sup>
 | {{yes}}
 | {{no}}<sup>21</sup>
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}<sup>7</sup>
 | {{yes|designated}}
 | {{yes}}<sup>24</sup>
 | {{yes}}<sup>24</sup>
 |-
 | path<sup>25</sup>
 | {{yes}}
 | {{no}}<sup>21</sup>
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}<sup>7</sup>
 | {{no}}<sup>11</sup>
 | {{yes}}<sup>25</sup>
 | {{yes}}<sup>25</sup>
 |}

# According to [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm#art29 Transit Code in Article 29], emergency vehicles in service (with siren lights turned on) have priority and have ''livre circulação'' (free movement). This expression also applied to regular vehicles on highways and is not understood by Brazilian jurists as meaning legal permission to use any kind of way in any situation at will. Using ways not meant for regular vehicles is discouraged by authorities and considered by emergency workers only if required by the circumstances.
# The [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm#anexoi Transit Code in Annex I] allows the following correspondences:
#* ''Ciclomotor'' ({{TagKey|moped}}): a low powered motorcycle capable of 50 km/h; distinct from a low powered bicycle ({{TagKey|mofa}})
#* ''Charrete'' ({{TagKey|carriage}}): an animal-powered vehicle for passenger transport; most Brazilian jurists consider ''montaria'' ({{TagKey|horse}} riding) a particular case
#* ''Via de trânsito rápido'' ({{TagKey|motorway}}): controlled-access freeway, with no at-grade intersections, no direct access to properties and no at-grade pedestrian crossings
# According to the [http://www.lex.com.br/legis_25174467_RESOLUCAO_N_465_DE_27_DE_NOVEMBRO_DE_2013.aspx National Transit Council in Resolution 465], low powered bicycles ({{TagKey|mofa}}) capable of 25 km/h are subject to the same rules of regular bicycles.
# Inline skates and skateboards are absent in national law, so they are considered pedestrians. Some municipalities allow them on cycleways.
# Motorways are designed for high-speed motorized traffic by [[Motorway|definition]].
# According to the [https://www.denatran.gov.br/images/Resolucoes/Resolucao5872016.pdf National Transit Council in Resolution 587], licensed agricultural vehicles within certain specifications are allowed on ''rodovias'' (paved rural highways). No national law prohibits them on ''vias de trânsito rápido'' ({{TagKey|motorway}}). [http://www.ctbdigital.com.br/artigo-comentarista/337] According to the [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm#art62 Transit Code in Article 62] and the [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm#art219 Transit Code in Article 219], vehicles must be capable of driving at half the speed limit. Therefore, {{TagKey|minspeed}} must be mapped, restricting access by most agricultural vehicles in practice.
# According to the [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm#art57 Transit Code in Article 57], ''ciclomotores'' ({{TagKey|moped}}) are subject to the rules that apply to motorcycles but are notably forbidden on ''vias de trânsito rápido'' ({{TagKey|motorway}})
# According to the [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm#anexoi Transit Code in Annex I], ''ciclo'' is a vehicle category that includes {{TagKey|bicycle}} and {{TagKey|moped}} (''ciclomotores''). According to the [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm#art244 Transit Code in Article 244], ''ciclos'' are forbidden on ''vias de trânsito rápido'' ({{TagKey|motorway}}) and ''rodovias'' (paved rural highways) except when they provide a {{TagKey|shoulder}}.
# According to the [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm#art68 Transit Code in Article 68], pedestrians may walk against vehicles even in paved rural highways without a shoulder. The law mentions this is forbidden in dangerous situations, without specifying them. Since bicycles are forbidden when a shoulder is absent, it could be argued that its absence represents a dangerous situation.
# Most types of motorized highways are not assigned any official national/administrative class by the Brazilian community, thus they are generally considered public roads/streets and have no special access restrictions other than present signage. Some restrictions apply outside urban areas, but it is not related to highway class in OSM.
# According to the [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm#art52 Transit Code in Article 52], animal-powered traffic carrying either cargo or passengers is subject to the rules of motorized traffic.
# According to the [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm#art58 Transit Code in Article 58], bicycles can ride on the sides of the motorized tracks of all public ways, except when a parallel separate cycleway is present, in which case the main way should have {{Key|bicycle|use_sidepath}}.
# Living streets are absent in national law. Municipal law may define some streets as {{Wikipedia|en|shared space|text=no}}, but currently none restrict the passage of any kind of vehicle, leaving restrictions to present signage. [https://forum.openstreetmap.org/viewtopic.php?id=64274 Some municipalities] confusingly use ''shared space'' to refer to shared bicycle lanes on motorized tracks or to paths that combine pedestrians and cyclists, none of which correspond to the concept of {{Wikipedia|en|living street|text=no}} (shared space between pedestrians, cyclists and motorized vehicles). Some believe that all Brazilian streets are in practice living streets, but motorists are liable only if a pedestrian has strictly followed the law, as stated in the [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm#art29§2 Transit Code in Article 29(2)]. According to [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm#capituloiv Transit Code in Chapter IV] and the [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm#art254 Transit Code in Article 254], pedestrians and cyclists have right of way if moving on the borders of motorized tracks when a sidewalk is unavailable or unusable, and when crossing on marked crossings or on traffic lights (they may cross at other points but do not have right of way and may be liable in some circumstances), leaving most of the area of motorized ways of any kind under right of way for motorized vehicles.
# Service ways are usually for access to amenities open for customers, thus usually not intended for passing public traffic. Access is not forbidden by law unless a barrier (such as a fence or wall) or signage is present. Segregated busways should be tagged with {{tag|service|driveway}}+{{tag|access|no}}+{{tag|emergency|yes}} and either {{tag|psv|designated}} if taxis and minibuses are allowed, or {{tag|bus|designated}} otherwise.
# Service ways usually occur inside gated private properties. When that's not the case, add {{tag|emergency|discouraged}}.
# Tracks are usually located in private rural properties or forests managed by the government, usually not intended for passing public traffic, even for agricultural or forestry purposes ({{tag|access|agricultural|;=forestry}}) or using agricultural vehicles ({{key|agricultural}}). Accidental access is usually not punished when no barrier (such as a fence or wall) and no signage is present to indicate that the way is privately owned.
# Tracks usually occur in areas without any barriers to access (farms, open nature). When they occur in areas managed by the government (forests) restricted by a barrier, add {{tag|emergency|destination}}.
# Usually wide enough for emergency vehicles.
# According to [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm#art68 Transit Code in Article 68], cyclists are considered pedestrians when pushing instead of riding, thus dismounted bicycles are always allowed on ways intended for pedestrians.
# Foot ways and pedestrian ways are mostly not signed, thus not designated. They are implicit from signs forbidding motorized vehicles or various types of construction and context.
# Rarely wide enough for emergency vehicles, should be tagged with {{tag|emergency|destination}} when it is physically possible.
# Since pedestrian ways are usually not signed, cycleways only exist legally when signage is present. In parks, according to the definition of ''logradouro público'' (public place) in the [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm#anexoi Transit Code in Annex I], unless signage is present, the space is primarily intended for pedestrians and not for vehicles of any kind, and according to the [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm#art96 Transit Code in Article 96], bicycles are considered vehicles, thus forbidden except if dismounted.
# According to the definitions of cycleway and cycle lane in the [http://www.planalto.gov.br/ccivil_03/leis/L9503.htm#anexoi Transit Code in Annex I], cycleways are exclusive to bicycles, thus pedestrians cannot use them by default. Allowed pedestrian access should be represented by adding {{tag|foot|designated}} when indicated by signs or {{tag|foot|yes}} when allowed by municipal law.
# Bridleways are absent in national law and very rare, so restrictions only apply if signage is present. Not the same as horse racing tracks, which are more common and should be tagged with {{tag|leisure|track}}+{{tag|sport|equestrian|;=horse_racing}}.
# Intentional footways are designated by the planner/maintainer, and {{tag|highway|path}} should be used when the intention is not clear (eg. it is an unofficial, unmaintained trail) or when the path is used for more than one equally important function (except when [[Bicycle#Miscellaneous|combined with cycleways]]). In these cases, restrictions for each type of vehicle should be explicitly tagged.

==== Denmark ====

Default access restrictions according to the Danish traffic regulations ("Færdselsloven" and "Bekendtgørelse om Vejafmærkning")
{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

 |-
 | motorway
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | trunk
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | primary

secondary

tertiary

unclassified

residential

living_street

 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | track (3)
 | {{no|no(4)}}
 | {{no|no(4)}}
 | {{no|no(4)}}
 | {{no|no(4)}}
 | {{no|no(4)}}
 | {{no|no(4)}}
 | {{partial|(5)}}
 | {{yes|yes(7)}}
 | {{yes|yes(7)}}

 |-
 | path (3)
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no|no(4)}}
 | {{partial|(5)}}
 | {{yes|yes(6)(7)}}
 | {{yes|yes(7)}}

 |-
 | bridleway
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no|no(1)}}
 | {{no}}
 | {{yes|designated}}
 | {{yes|yes(2)}}

 |-
 | footway 

pedestrian

 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}

 |}

# The regulations distinguish between mofa ("lille knallert") having a maximum speed of 30 km/h and moped ("stor knallert") having a maximum speed of 45 km/h. Mofas are allowed on cycleways.
# Pedestrians may only use cycleways if there are no adjacent footways. For routing purposes it is safe to assume that pedestrian access is permitted.
# Rules for 'path' and 'track' based on regulations for access to forests and countryside ([http://naturstyrelsen.dk/naturoplevelser/regler-i-naturen/hvor-maa-jeg-faerdes/ Vejledning om adgangsreglerne])
# Motor vehicles including mopeds and mofas are not allowed in forests. Motor vehicles are allowed on tracks and paths in the countryside if the track or path appears suitable for the type of vehicle. The landowner may however prohibit motorized access using signs.
# Horses are not allowed in privately owned forests unless the owner has explicitly permitted horse riding. In state owned forests horses are allowed on paved tracks and on unpaved tracks wider than 2.5 meter. In the countryside horses are allowed on suitable tracks and paths.
# Cycling is allowed on paths that appear suitable for normal bicycles.
# Access to private forests is only permitted between 6:00 and sunset.

==== France ====

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

 |-
 | motorway

trunk
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | primary

secondary

tertiary

unclassified

residential

track

 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 |living_street
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | path 
 | {{no}}
 | {{partial|tbd (2)}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{no}}
 |-
 | footway 

 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}

 |-
 | pedestrian

 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{partial|yes (1)}}
 | {{yes|designated}}

 |}

# Their speed is limited to pedestrian speed
# to be defined (law is unclear)

==== Finland ====

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

 |-
 | motorway
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | trunk
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | primary

secondary

tertiary

unclassified

residential

 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 |living_street
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 |track
 | {{partial|yes (5)}}
 | {{partial|yes (5)}}
 | {{partial|yes (5)}}
 | {{partial|yes (5)}}
 | {{partial|yes (5)}}
 | {{partial|yes (5)}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | path 
 | {{no|private}}
 | {{no|private}}
 | {{no|private}}
 | {{no|private}}
 | {{no|private}}
 | {{no|private}}
 | {{partial|(1)}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{partial|yes (2)}}
 | {{no}}
 | {{yes|designated}}
 | {{partial|yes (3)}}
 |-
 | footway 

 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}

 |-
 | pedestrian

 | {{partial|(4)}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes|designated}}

 |}

Some trunk roads ban pedestrians and cyclists but they are individually tagged.

# Horses not allowed on any designated footways or cycleways.
# Allowed only when additional sign "Sallittu mopoille" is present.
# Should use nearby footway if present. Most often is specifically designated for both, though, and tagged as such. If only a cycleway is present next to a road without sidewalks (never seen such a case), pedestrians "should" use the cycleway.
# Legal pedestrian road ("kävelykatu", liikennemerkki nr. 575) allows motorcars when it's the ''only'' way to premises - other roads tagged as pedestrian can allow with extra signs and might be off limits to cyclists. The local default for tagging has not been agreed upon.
# For privately owned tracks, only non-continous use and up to minor wear allowed without permission or participation in the road costs. In addition, may be restricted by the private road owner but it is not possible to restrict access if the road has been subsidiced recently with (a particular form of) public funding.

==== Germany ====

Default access restrictions if not tagged on the individual street.

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| mofa
!| horse
!| bicycle
!| foot

 |-
 | motorway and streets with key motorroad=yes
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | trunk 
primary
secondary
tertiary

unclassified

residential

living_street

road
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | service 
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | path 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{partial|''yes'' (1)}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{partial|''yes'' (2)}}
 | {{no}}
 | {{yes|designated}}
 | {{no}}

 |-
 | footway 
pedestrian
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no|dismount}} (3)
 | {{yes|designated}}

 |-
 | track (proposal)
 | {{yes|destination}}
 | {{yes|destination}}
 | {{yes|destination}}
 | {{yes|destination}}
 | {{yes|destination}}
 | {{yes|destination}}
 | {{yes|destination}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |}

# Only with {{tag|bicycle|designated}} in extra urban areas. ( StVO §2.4: ''Außerhalb geschlossener Ortschaften dürfen Mofas Radwege benutzen.'' )
# Only in extra urban areas. ( StVO §2.4: ''Außerhalb geschlossener Ortschaften dürfen Mofas Radwege benutzen.'' )
# Laut StVO dürfen Fahrräder auf Fußwegen grundsätzlich geschoben werden, wenn damit andere Fußgänger nicht erheblich behindert werden.

==== Hong Kong ====

Default access restrictions for Hong Kong. ''Italic'' means inherited from parent category.

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| access
!| motor_vehicle
!| motorcycle
!| motorcar
!| goods
!| hgv
!| psv
!| share_taxi
!| horse
!| bicycle
!| foot

 |-
 | motorway
 | {{no|no}}
 | {{yes|designated}}
 | {{yes|''designated''}} (2)
 | {{yes|''designated''}}
 | {{yes|''designated''}}
 | {{yes|''designated''}}
 | {{yes|''designated''}}
 | {{no|no}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}

 |-
 | trunk

primary

secondary

tertiary

unclassified

residential

living_street

road

 | {{yes}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{partial|''yes'' (1)}}
 | {{yes|''yes''}}
 |-
 | pedestrian
 | {{no}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes}}

 |-
 | path 
 | {{no}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes}}
 | {{partial|''yes'' (1)}}
 | {{yes}}

 |-
 | cycleway 
 | {{no}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes|designated}}
 | {{no|''no''}}

 |-
 | footway 
 | {{no}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes|designated}}
 |}

(1) Bicycles must use a cycleway alongside if present.
(2) Motorcycles are allowed on motorways only if its displacement is at least 125 cc.

==== Hungary ====

Default access restrictions for Hungary.

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| access
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

 |-
 | motorway

trunk
 | {{yes}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | primary
 | {{yes}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{no}} (1)
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}

 |-
 | secondary

tertiary

unclassified

residential

road

 | {{yes}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}

 |-
 | living_street
 | {{partial|Destination}}
 | {{partial|''Destination''}}
 | {{partial|''Destination''}}
 | {{partial|''Destination''}}
 | {{partial|''Destination''}}
 | {{partial|''Destination''}}
 | {{partial|''Destination''}}
 | {{partial|''Destination''}}
 | {{yes}}
 | {{yes}}

 |-
 | pedestrian
 | {{no}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{yes}}

 |-
 | path 
 | {{no}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway 
 | {{no}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{yes|Designated}}
 | {{no|''No''}}
 | {{no|''No''}}

 |-
 | cycleway 
 | {{no}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{yes|Designated}}
 | {{no|''No''}} (1)

 |-
 | footway 
 | {{no}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{yes|Designated}}
 |}

# Allowed only if there are no alternatives.


==== Iceland ====

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot


 |-
 | trunk

primary

secondary

tertiary

unclassified

residential

living_street

road

 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | track
 | {{partial}}
 | {{partial}}
 | {{partial}}
 | {{partial}}
 | {{partial}}
 | {{partial}}
 | {{partial}}
 | {{yes}}
 | {{yes}}

 |-
 | path
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway (2)
 | {{partial}}
 | {{partial}}
 | {{partial}}
 | {{partial}}
 | {{partial}}
 | {{partial}}
 | {{yes|designated}}
 | {{partial}}
 | {{partial}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{yes|yes}}

 |-
 | footway 

pedestrian

 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes|designated}}

 |}

# 'Bridleway' does not exist as a legal term. They are either on private grounds or signposted. Its also possible that the local government has put down restrictions for an area.

==== Italy ====

Proposed access restrictions. Not yet endorsed by the Italian OSM community.

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| access
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

|-
 | path 
 | {{no}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | cycleway 
 | {{no}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes|designated}}
 | {{no|''no''}}

 |-
 | footway
pedestrian 
 | {{no}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|dismount}}
 | {{yes|designated}}

|-
 | track

unclassified

residential

service

tertiary

secondary

primary
 
 | {{yes}}
 | {{yes}} 
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

|-
 | trunk

motorway
 
 | {{no}}
 | {{yes}} 
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{no}}
 | {{no}}
 | {{no}}
 |}

==== Netherlands ====

Default access restrictions according to the traffic regulations in the Netherlands ([https://wetten.overheid.nl/BWBR0004825/2019-07-01 Reglement verkeersregels en verkeerstekens 1990 (RVV 1990) as in force on 2008-10-07])

For a nice overview see the PDF in [https://staticresources.rijkswaterstaat.nl/binaries/Boekje%20Verkeersborden%20en%20verkeersregels%20in%20Nederland_tcm174-337518_tcm21-9388.pdf Dutch] / [https://staticresources.rijkswaterstaat.nl/binaries/Booklet%20Road%20traffic%20signs%20and%20regulations%20in%20the%20Netherlands_tcm174-337519_tcm21-9389.pdf English].
{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| mofa
!| horse
!| bicycle
!| foot

 |-
 | motorway
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | trunk
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | primary<br/> secondary<br/> tertiary<br/> unclassified<br/> residential<br/> living_street<br/> road <br/>track (1)
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | path (2)
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{no}}
 | {{partial|(3)}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no|no (4)}}
 | {{partial|(5)}}
 | {{no}}
 | {{yes|designated}}
 | {{yes|yes (6)}}

 |-
 | footway<br/> pedestrian
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}

 |}

# Access on tracks is not restricted unless prohibited by some sign or barrier. Note however that a lot of tracks in forests that having signs for "access=no foot=permissive" are not (yet) tagged with access tags.
# The regulations define a lane ("rijbaan") as any part of a road that is suitable for motor vehicles ("rijdende voertuigen").
# Pedestrians are not supposed to use a bridleway ("ruiterpad"), when there is a road next to it. The regulations seem silent whether pedestrians can use a bridleway if no other alternative is available.
# Mopeds ("bromfietsen") are not allowed to use cycleways indicated by the compulsory cycleway sign G11, or the optional cycleway sign G13. There is however a sign that allows both bicycles and mopeds (G12a).
# Mofas ("snorfietsen" < 25km/u) must use cycleways indicated by the compulsory cycleway signs G11 and G12a, but are not allowed to use cycleways indicated by the optional cycleway sign G13.
# Pedestrians ("voetgangers") are allowed to use cycleways if no parallel footpath is available, which is mostly missing outside of urban areas.
<gallery perrow="3">
File:Bord g11.png|alt=G11: Round blue sign with a white bycicle|G11: Compulsory Cycleway
File:Bord g12a.png|alt=G12a: Round blue sign with a white moped and bycicle|G12a: Compulsory Moped and Cycleway
File:Bord g13.png|alt=G13: Rectangular blue sign with white text 'fietspad'|G13: Optional Cycleway
</gallery>

==== Norway ====

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

 |-
 | motorway (1)

highway with key motorroad=yes (1)

 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | trunk

primary

secondary

tertiary

unclassified

residential

living_street

road

 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | service 
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | track (3)
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{partial}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | path
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway (2)
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{partial}}
 | {{partial}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{yes|yes}}

 |-
 | footway 

pedestrian

 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes|designated}}

 |}

# Motorways and motorroads are only for motor vehicles constructed to move at least 40 km/h on flat ground.
# 'Bridleway' does not exist as a legal term. They are either on private grounds or signposted. Its also possible that the local government has put down restrictions for an area.
# Tracktype 1 and 2 can be considered, but tracks are mostly for agricultural use.

==== Oman ====

Proposed default access restrictions for Oman. Not endorsed by the Oman OSM community (yet).

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| access
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle 
!| foot

 |-
 | motorway
 | {{yes}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}
 | {{yes}} (1)
 | {{no}}

 |-
 | trunk
 | {{yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}

 |-
 | primary

secondary

tertiary

unclassified

residential

living_street

road

 | {{yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 | {{yes|yes}}
 |-
 | pedestrian
 | {{no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{yes|designated}}

 |-
 | path 
 | {{no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway 
 | {{no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{yes|designated}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}

 |-
 | cycleway 
 | {{no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{yes|designated}}
 | {{yes|designated}}

 |-
 | footway 
 | {{no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{no|no}}
 | {{yes|designated}}
 |}

# Cycling is allowed in Oman on every road, including motorways.

====Philippines====
Default access restrictions for the Philippines according to the Land Transportation and Traffic Code (Republic Act 4136) and supplementing laws (e.g. Limited Access Highway Act). Not yet endorsed by the Philippine OSM community

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| access
!| motorcar
!| motorcycle 
!| goods 

hgv
(1)

!|psv (2)
!| moped (3)
!| horse
!| bicycle
!| foot

 |-
 | motorway (4)
motorway_link (4)
 | {{yes}}
 | {{yes|Designated}} (3)
 | {{yes|Designated}}
 | {{yes|Designated}}
 | {{yes|Designated}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}


 |-
 | trunk

trunk_link

primary

primary_link

secondary

secondary_link

tertiary

tertiary_link

unclassified

residential

living_street

road

 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}} (5)
 | {{yes}} (5)
 |-
 | pedestrian
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes}}

 |-
 | path 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway (6)
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|Designated}}
 | {{yes}}
 | {{yes}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|Designated}}
 | {{yes}}

 |-
 | footway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|Designated}}
 |}

# There is no distinction between heavy and light trucks. The no trucks sign implies a ban on all trucks (except deliveries) unless otherwise specified.
# "Public utility" vehicles (for hire and/or carrying more that 9 persons), including jeepneys and tricycles (motorized rickshaws).
# Legally, only for those with an engine displacement of 400cc. Mopeds are treated the same as motorcycles.
# Only vehicles which can attain a top speed of 60 km/h are allowed.
# Pedestrians are allowed by default on any road that are not expressways, unless signed otherwise. There are no special laws on bicycles, but they can be assumed to follow the same rules as pedestrians.
# Laws do not mention anything about horse trails. 

==== Poland ====

Default access restrictions for Poland.

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| access
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

 |-
 | motorway

trunk
 | {{yes}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{No}}
 | {{No}}
 | {{No}}
 | {{No}}

 |-
 | primary
 | {{yes}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{no}} (1)
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}

 |-
 | secondary

tertiary

unclassified

residential

road

 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | living_street
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | pedestrian
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}

 |-
 | path 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{no}} (1)

 |-
 | footway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 |}

# Allowed only if there are no alternatives.

==== Romania ====

Default access restrictions if not tagged on the individual street.

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

 |-
 | motorway
trunk
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | primary
secondary

tertiary

unclassified

residential

road

track
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | living_street
 | destination
 | destination
 | destination
 | destination
 | destination
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | path 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{no|No (1)}}

 |-
 | footway 
pedestrian
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}

 |}

# Pedestrians allowed if a sidewalk or a dedicated footpath is missing.

==== Russia ====

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| access
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

 |-
 | motorway
 | {{yes|''yes''}}
 | {{yes|''designated''}}
 | {{yes|''designated''}}
 | {{yes|''designated''}}
 | {{yes|''designated''}}
 | {{yes|''designated''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}

 |-
 | trunk
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{partial|(1)}}
 | {{partial|(2)}}

 |-
 | primary

secondary

tertiary

unclassified

residential

 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{partial|(1)}}
 | {{partial|(2)}}
 |-
 | living_street
 | {{yes}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{yes|''yes''}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 |-
 | service
 | {{yes}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{yes|''yes''}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 |-
 | pedestrian
 | {{yes|''yes''}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{yes}}

 |-
 | path 
 | {{yes|''yes''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway 
 | {{yes|''yes''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes|designated}}
 | {{no|''no''}}
 | {{no|''no''}}

 |-
 | cycleway 
 | {{yes|''yes''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes}}
 | {{no|''no''}}
 | {{yes|designated}}
 | {{partial|(4)}}
 |-
 | footway 
 | {{yes|''yes''}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{yes|designated}}
 |-
 | footway + footway=sidewalk
 | {{yes|''yes''}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{yes|designated}}
 |-
 | footway + footway=crossing
 | {{yes|''yes''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes|designated}}
 |-
 | steps
 | {{yes|''yes''}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{partial|(3)}}
 | {{yes|designated}}
 |-
 | road
 | {{partial|(5)}}
 | {{partial|(5)}}
 | {{partial|(5)}}
 | {{partial|(5)}}
 | {{partial|(5)}}
 | {{partial|(5)}}
 | {{partial|(5)}}
 | {{partial|(5)}}
 | {{partial|(5)}}
 | {{partial|(5)}}
 |}
# Mopeds and bicycles allowed if a cycleway is missing<ref><blockquote>[http://www.gai.ru/voditelskoe-udostoverenie/pddrf/#p24 23.3]<br />...<br />двигаться по дороге при наличии рядом велосипедной дорожки;</blockquote></ref>.
# Pedestrians allowed if a sidewalk and a footway are missing. Pedestrians, which have large cargo, and people, using whellchair without engine, can use side of carriageway, if their movement along sidewalk or roadside hinder other pedestrians<ref><blockquote>[http://www.gai.ru/voditelskoe-udostoverenie/pddrf/#p4 4.1.]<br />...<br />Пешеходы должны двигаться по тротуарам или пешеходным дорожкам, а при их отсутствии — по обочинам. Пешеходы, перевозящие или переносящие громоздкие предметы, а также лица, передвигающиеся в инвалидных колясках без двигателя, могут двигаться по краю проезжей части, если их движение по тротуарам или обочинам создает помехи для других пешеходов</blockquote></ref>.
# Vehicles of road's and communal's stuff allowed. Vehicles, which move cargo to objects, which located directly close to footway, allowed, if they use shortest way. Moving of vehicles should be safe<ref><blockquote>[http://www.gai.ru/voditelskoe-udostoverenie/pddrf/#p9 9.9]<br />...<br />Допускается движение машин дорожно-эксплуатационных и коммунальных служб, а также подъезд по кратчайшему пути транспортных средств, подвозящих грузы к торговым и другим предприятиям и объектам, расположенным непосредственно у обочин, тротуаров или пешеходных дорожек, при отсутствии других возможностей подъезда. При этом должна быть обеспечена безопасность движения.</blockquote></ref>. 
# Pedestrians allowed if a sidewalk, a footway and roadside are missing<ref><blockquote>[http://www.gai.ru/voditelskoe-udostoverenie/pddrf/#p4 4.1.]<br />...<br />При отсутствии тротуаров, пешеходных дорожек или обочин, а также в случае невозможности двигаться по ним пешеходы могут двигаться по велосипедной дорожке или идти в один ряд по краю проезжей части (на дорогах с разделительной полосой — по внешнему краю проезжей части).</blockquote></ref>.
# Common rule about routing along highway=road can not be determined, because highway=road is a line which can be a road or not, and if it is a road, it is a road with unknown status. The recommendation is to exclude highway=road from routing or to give such lines the lower status in roads' graph.

==== Slovakia ====

Default access restrictions for Slovakia.

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| access
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

 |-
 | motorway

trunk
 | {{yes}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | primary
 | {{yes}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{no}} (1)
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}

 |-
 | secondary

tertiary

unclassified

residential

road

 | {{yes}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}
 | {{yes|''Yes''}}

 |-
 | living_street
 | {{partial|Destination}}
 | {{partial|''Destination''}}
 | {{partial|''Destination''}}
 | {{partial|''Destination''}}
 | {{partial|''Destination''}}
 | {{partial|''Destination''}}
 | {{partial|''Destination''}}
 | {{partial|''Destination''}}
 | {{yes}}
 | {{yes}}

 |-
 | pedestrian
 | {{no}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{yes}}

 |-
 | path 
 | {{no}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway 
 | {{no}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{yes|Designated}}
 | {{no|''No''}}
 | {{no|''No''}}

 |-
 | cycleway 
 | {{no}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{yes|Designated}}
 | {{no|''No''}} (1)

 |-
 | footway 
 | {{no}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{no|''No''}}
 | {{yes|Designated}}
 |}

# Allowed only if there are no alternatives.

==== Spain ====

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

 |-
 | motorway

trunk (motorroad = yes)
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | trunk (no motorroad)

primary

secondary

tertiary

unclassified

residential

 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{partial|yes (1)}}
 | {{yes}}
 | {{partial|yes (2)}}

 |-
 | track
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 |living_street
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | path 
 | {{no}}
 | {{partial|¿?}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}
 | {{no}}
 |-
 | footway 

 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|designated}}

 |-
 | pedestrian

 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{partial|yes (3)}}
 | {{yes|designated}}

 |}

# Only if there is no alternative track. The highway of lowest intensity of traffic must be selected
# Use the left side
# Their speed is limited to pedestrian speed

==== Switzerland ====

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| mofa
!| horse
!| bicycle
!| inline_skates (6)
!| foot
 |-
 | motorway
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | trunk
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{yes|designated}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}} 
 | {{no}}

 |-
 | primary

secondary

tertiary

unclassified

residential

living_street

road

track

 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes| Yes (7)}}
 | {{yes}}
 |-

 | pedestrian
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no|No (1)}}
 | {{yes}}
 | {{yes}}

 |-
 | path 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no|No (5)}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|Designated}}
 | {{no|No (2)}}
 | {{no|No (2)}}
 | {{no|No (2)}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}} 
 | {{yes|Yes (2)}}
 | {{no|No (3)}}
 | {{yes|Designated}}
 | {{yes}}
 | {{partial|No (4)}}

 |-
 | footway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no|No (2)}}
 | {{no|No (2)}}
 | {{yes}}
 | {{yes|Designated}}
 |}
# (de) http://www.admin.ch/ch/d/sr/741_21/a22c.html, (fr) http://www.admin.ch/ch/f/rs/741_21/a22c.html
# (de) http://www.admin.ch/ch/d/sr/741_21/a33.html, (fr) http://www.admin.ch/ch/f/rs/741_21/a33.html
# (de) http://www.admin.ch/ch/d/sr/741_01/a43.html, (fr) http://www.admin.ch/ch/f/rs/741_01/a43.html
# Pedestrians allowed if a sidewalk or a foot way is missing (de) http://www.admin.ch/ch/d/sr/741_11/a40.html, (fr) http://www.admin.ch/ch/f/rs/741_11/a40.html
# tbd
# Swiss road regulations have a seperate category for inline skates, skates, kick boards and similar devices: "fahrzeugähnliche Geräte". (de) http://www.admin.ch/ch/d/sr/741_11/a1.html, (fr) http://www.admin.ch/ch/f/rs/741_11/a1.html
# If a side walk is present it has to be used, except in residential zones with a speed limit of 30 km/h or on living streets (de) http://www.admin.ch/ch/d/sr/741_11/a50.html, (fr) http://www.admin.ch/ch/f/rs/741_11/a50.html

====Turkey====
{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| access
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

 |-
 | motorway
 | {{yes}}
 | {{yes|''designated''}}
 | {{yes|''designated''}}
 | {{yes|''designated''}}
 | {{yes|''designated''}}
 | {{yes|''designated''}}
 | {{no|no(1)}}
 | {{no}}
 | {{no|no(1)}}
 | {{no|no(1)}}

 |-
 | trunk
 | {{yes}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}

 |-
 | primary

secondary

tertiary

unclassified

residential

living_street

road

 | {{yes}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 |-
 | pedestrian
 | {{no}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes}}

 |-
 | path 
 | {{no}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway 
 | {{no}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes|destination}}
 | {{no|''no''}}
 | {{yes}}

 |-
 | cycleway 
 | {{no}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes|designated}}
 | {{partial|(2)}}

 |-
 | footway 
 | {{no}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes|designated}}
 |}

# Only if no other alternatives are present, "yes".
# If the signs "footway" and "cycleway" are posted together, or if there is no warning "no foot", foots can use cycleways.

==== Ukraine ====

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| access
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

 |-
 | motorway

trunk with motorroad=yes
 | {{yes|''yes''}}
 | {{yes|''designated''}}
 | {{yes|''designated''}}
 | {{yes|''designated''}}
 | {{yes|''designated''}}
 | {{yes|''designated''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 |-
 | trunk

primary

secondary

tertiary

unclassified

residential

track

road

 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 | {{yes|''yes''}}
 |-
 | living_street
 | {{yes}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{yes|''yes''}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{yes|''yes''}}
 |-
 | service
 | {{yes}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{yes|''yes''}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{partial|''destination''}}
 | {{yes|''yes''}}
 |-
 | pedestrian
 | {{yes|''yes''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{partial|''dismount''}}
 | {{yes}}

 |-
 | path 
 | {{yes|''yes''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway 
 | {{yes|''yes''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes|designated}}
 | {{no|''no''}}
 | {{no|''no''}}

 |-
 | cycleway 
 | {{yes|''yes''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes|designated}}
 | {{no|''no''}}
 |-
 | footway 
 | {{yes|''yes''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{partial|''dismount''}}
 | {{yes|designated}}
 |-
 | steps
 | {{yes|''yes''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{no|''no''}}
 | {{yes|designated}}
 |}

====United Kingdom====


{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

 |-
 | motorway
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | trunk

primary

secondary

tertiary

unclassified

residential

 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 |living_street
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | path 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}*
 | {{yes}}*
 | {{yes}}*

 |-
 | bridleway
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes}}
 |-
 | footway 

 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}

 |-
 | pedestrian

 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}

 |}

<nowiki>*</nowiki> : The [[United Kingdom Tagging Guidelines]] state that highway=path, when used it the UK, implies "a generic narrow path that is used in conjunction with access tags". This makes the default "yes" assumption dubious.

The legal status of cycles on public rights of way in England and Wales is disputed. On pavements, it is illegal to cycle unless designated (hence not a default setting here) but on public footpaths the law is less clear.

In Scotland, there is a right of access by any non-motorised transport (including foot, bicycle and horse) just about everywhere (except for some specific exclusions).

====United States of America====
Note: Regulations vary significantly by state.

{| class="wikitable" border="1" cellspacing="0" cellpadding="4"
 |- style="background-color:#E9E9E9"
!| {{Tag|highway}}
!| motorcar
!| motorcycle
!| goods
!| hgv
!| psv
!| moped
!| horse
!| bicycle
!| foot

 |-
 | motorway
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}

 |-
 | trunk

primary

secondary

tertiary

unclassified

residential

living_street

road

 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 |-
 | pedestrian
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes}}
 |-
 | path
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes}}
 | {{yes}}
 | {{yes}}
 | {{yes}}

 |-
 | bridleway (1)
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|Designated}}
 | {{partial|Yes}}
 | {{partial|Yes}}

 |-
 | cycleway 
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|Designated}}
 | {{yes}}

 |-
 | footway
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{no}}
 | {{yes|Designated}}
 |}

# I couldn't find any specific laws regarding non-horse use of horse trails.  They don't seem to be an official designation of the national gov't.

==Tools using this information==

????

== References ==
<references />

<!-- Categories -->
[[Category:Tags]]
[[Category:Routing]]
"""
```
